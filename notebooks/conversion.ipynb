```python
data_path = "./../data/raw/ADNI"
subjects = os.listdir(data_path)
```

```python
session_paths = []
for subject in subjects:
    string = "3-plane_localizer"
    localizer_path = os.path.join(data_path,subject,string)
    if os.path.exists(localizer_path):
        dates_paths = [os.path.join(localizer_path,pth) for pth in os.listdir(localizer_path) if pth != '.DS_Store']
        for date in dates_paths:
            temp = os.listdir(date)[0] #print(os.listdir(os.path.join(date,temp)))
            session_paths.append(os.path.join(date,temp))
```
